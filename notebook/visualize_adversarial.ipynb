```python
clst = misc.SegmentedSerializableList(
    inclconf.PROJECT_DIRECTORY
    / "data"
    / "datasets"
    / "CIFAR10Adversarial"
    / "cw"
    / "densenet121",
    "train",
)

cx, cy = clst[1]
plt.title(f"category: {cy}")
plt.axis("off")
plt.imshow(cx)
plt.show()
```
